Let's try it again without re-encoding each frame. I hypothesize that this should give us more consistency from frame-to-frame (since we're using manipulating a single latent representations throughout the process, whereas re-encoding would give us an indpendent latent each time... maybe there's something I can do about that?), but at the cost of losing coherence with the video.

````python
%%writefile config/conf/spring2winter_noreencode.yaml
# @package _global_
animation_mode: Video Source
video_path: /home/dmarx/proj/pytti-book/lfb_trimmed2.mp4

image_model: VQGAN
vqgan_model: imagenet

scenes: "a photograph of a beautiful spring day:2 ||  a photograph of a snowy winter day:2 "
#cut_pow: 1.5
cutouts: 60
#gradient_accumulation_steps: 2

reencode_each_frame: false
#reencode_each_frame: true

steps_per_scene: 3840
interpolation_steps: 960

steps_per_frame: 80
save_every: 80 
display_every: 1000000000000000
direct_stabilization_weight: 1
#edge_stabilization_weight: 1
#flow_stabilization_weight: 1
#depth_stabilization_weight: 1

backups: 3

file_namespace: spring2winter_noreencode #video_lfb_4s_spring2winter_noreencode

#RN50: true
ViTB32: true

use_tensorboard: true
````
